### **Handling Missing Value 1.5 (ลบค่า missing เฉพาะใน column ที่เราสนใจ)**

```python
natural_nomissingAge = natural.dropna(subset=['ตำบล'])
```

```python
removed2 = natural.shape[0] - natural_nomissingAge.shape[0]
```

```python
print(f'size before drop = {natural.shape[0]}')
print(f'size after drop = {natural_nomissingAge.shape[0]}')
print(f'We lost {100*(removed2/natural.shape[0])}% of data')
```

**ใส่ในช่องว่างข้อมูลแทนตรง NaN เพราะว่าการ dropna ทำให้ข้อมูลหายไปเยอะเกินไป**

```python
natural[['ตำบล']] = natural[['ตำบล']].fillna('ไม่มีข้อมูล')
```
