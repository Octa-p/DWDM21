## Scatter plot

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
list(range(len(data.iloc[-20:,1])))
```

```python
data.iloc[-20:,1] #ค่า positive case
```

```python
plt.scatter(range(len(data.iloc[-20:,1])),data.iloc[-20:,1]) #ใช้ค่า defult
```

```python
plt.scatter(range(len(data.iloc[-20:,1])),data.iloc[-20:,1],s=100) #ทำให้ตัว marker ใหญ่ขึ้น ในที่นี้คือ วงกลม
```

```python
plt.scatter(range(len(data.iloc[::50,1])),data.iloc[::50,1],s=0.03*data.iloc[::50,2]) #ตั้งแต่วันแรกจนถึงวันสุดท้าย กระโดดมาวันละ 50 วัน #ขนาดของวงคือตรวจได้เท่าไหร่ต่อวัน ความสูงคือเจอกี่เคส
```

```python
plt.scatter(range(len(data.iloc[::50,1])),data.iloc[::50,1],s=0.03*data.iloc[::50,2],c='r') # r g b  c m y k w
```

```python
plt.scatter(range(len(data.iloc[::50,1])),data.iloc[::50,1],s=0.03*data.iloc[::50,2],c=['r','r','r','g','g','g','b','b','b','c','m','k']) #สามารถกำหนดสีแต่ละจุดได้
```

```python
plt.scatter(range(len(data.iloc[::50,1])),data.iloc[::50,1],s=0.03*data.iloc[::50,2],c=[[1.,0.5,0]]) #rgb [[r,g,b]] r,g,b = [0-1] 0=ไม่มีสี 1 = เข้มสุด #เอาสีผสมกัน
```

```python
plt.scatter(range(len(data.iloc[::50,1])),data.iloc[::50,1],s=0.03*data.iloc[::50,2],c=[[1.,0.2,0]]) #rgb
```

```python
plt.scatter(range(len(data.iloc[::50,1])),data.iloc[::50,1],s=300,c=0.03*data.iloc[::50,2]) ##ใช้ค่าตัวเลข สีเข้ม = ตรวจน้อย สีสว่าง = ตรวจมาก
```
